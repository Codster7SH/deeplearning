```python
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
image = mpl.image.imread("/kaggle/input/imagesegmentation/wp11614129-volkswagen-virtus-wallpapers.jpg")
plt.imshow(image)
```

```python
image.shape
```

here 478 & 579 tells the pixel , while 4 tell about the colours/dimensions

```python
X = image.reshape(-1,3)
```

```python
X.shape
```

we have splitted the 2 co-ordinates into pixels and kept, the othr one constatant

```python
kmeans = KMeans(n_clusters=5, n_init=10) #we gave 2 as a defined value else it will show error , more the clusters more will be the details
```

```python
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img / 255)   #div. with 255 done to get the value between 0 and 1
```

```python
import cv2

cv2.imwrite("image1.png", cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) #as opencv uses BGR colour scheme and we want to use rgb to we have swapped 
cv2.imwrite("image1.png", cv2.cvtColor(segmented_img.astype("uint8"), cv2.COLOR_BGR2RGB)) 
```

**Let's do it for different Image****

```python
image2 = mpl.image.imread("/kaggle/input/imagesegmentation/jamie-street-UtrE5DcgEyg-unsplash.jpg")
plt.imshow(image2)
```

```python
image2.shape
```

```python
Y = image2.reshape(-1,3)
```

```python
Y.shape   #this image is of good quality so, has much pixels
```

```python
kmeans = KMeans(n_clusters=3, n_init=10) #we gave 2 as a defined value else it will show error , more the clusters more will be the details
```

```python
kmeans.fit(Y)
```

```python
segmented_img2 = kmeans.cluster_centers_[kmeans.labels_]
segmented_img2 = segmented_img2.reshape(image2.shape)
```

```python
plt.imshow(segmented_img2 / 255)
```

Here the cluster were less, that's why image is less detailed

```python
cv2.imwrite("image3.png", cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
cv2.imwrite("image3.png", cv2.cvtColor(segmented_img2.astype("uint8"), cv2.COLOR_BGR2RGB))
```
